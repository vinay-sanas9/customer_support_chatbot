```python
import textblob
# TextBlob is used for sentiment analysis
from textblob import TextBlob

# Counter helps count repeated queries
from collections import Counter

# re is used for text cleaning using regular expressions
import re

# CHATBOT CLASS
class ChatBot:

    # Constructor Method
    # This runs automatically when the chatbot starts
    def __init__(self):

        # Defining chatbot intents
        # Each intent contains:
        # keywords -> words user may type
        # response -> chatbot reply

        self.intents = {

            "greeting": {
                "keywords": ["hello", "hi", "hey", "good morning"],
                "response": "Hello! How can I help you today?"
            },

            "hours": {
                "keywords": ["hour", "hours", "open", "close", "timing"],
                "response": "We are open from 9 AM to 5 PM, Monday to Friday."
            },

            "return": {
                "keywords": ["refund", "moneyback", "return"],
                "response": "I can help you with your return request. Connecting you to a live agent."
            },

            "pricing": {
                "keywords": ["price", "cost", "pricing", "fees"],
                "response": "Please visit our pricing page for detailed plans and offers."
            },

            "support": {
                "keywords": ["help", "support", "issue", "problem"],
                "response": "Our support team will help you resolve your issue."
            },

            "thanks": {
                "keywords": ["thanks", "thank you"],
                "response": "You're welcome! Happy to help."
            }
        }

        # Conversation history
        # Stores all user messages

        self.history = []

        # Sentiment tracking

        self.positive_count = 0
        self.negative_count = 0
        self.neutral_count = 0

        # Query counter
        # Tracks most common queries

        self.query_counter = Counter()

    # TEXT PREPROCESSING FUNCTION

    def preprocess_text(self, text):

        # Convert text to lowercase
        text = text.lower()

        # Remove special characters and punctuation
        text = re.sub(r'[^a-zA-Z\s]', '', text)

        # Split sentence into words
        words = text.split()

        return words

    # RESPONSE GENERATION FUNCTION

    def get_response(self, message):

        # Save user message in history
        self.history.append(message)

        # Count user queries
        self.query_counter[message.lower()] += 1

        # Preprocess user message
        message_words = self.preprocess_text(message)

        # INTENT MATCHING

        for intent_name, intent_data in self.intents.items():

            # Check whether any keyword exists in user message
            if any(word in message_words for word in intent_data["keywords"]):

                return intent_data["response"]

        # SENTIMENT ANALYSIS

        # TextBlob calculates sentiment polarity
        # Positive = >0
        # Negative = <0
        # Neutral = 0

        sentiment = TextBlob(message).sentiment.polarity

        # Positive Sentiment
        if sentiment > 0:

            self.positive_count += 1

            return "That's great to hear!"

        # Negative Sentiment
        elif sentiment < 0:

            self.negative_count += 1

            return "I'm sorry to hear that. How can I help you?"

        # Neutral Sentiment
        else:

            self.neutral_count += 1

            return "I see. Can you tell me more about that?"

    # CHATBOT ANALYTICS FUNCTION

    def show_analytics(self):

        print("\n================ CHAT ANALYTICS ================\n")

        # Total messages
        print(f"Total Messages: {len(self.history)}")

        # Sentiment statistics
        print(f"Positive Messages: {self.positive_count}")
        print(f"Negative Messages: {self.negative_count}")
        print(f"Neutral Messages: {self.neutral_count}")

        # Most common queries
        print("\nMost Common Queries:")

        for query, count in self.query_counter.most_common(5):

            print(f"{query} --> {count} times")

        print("\n================================================")


    # MAIN CHAT FUNCTION

    def chat(self):

        print("ChatBot: Hi! How can I help you today?")
        print("Type 'exit', 'quit', or 'bye' to end the chat.\n")

        # Infinite loop for conversation
        while True:

            try:

                # Taking user input
                user_message = input("You: ").strip()

                # Exit condition
                if user_message.lower() in ['exit', 'quit', 'bye']:

                    print("\nChatBot: Thank you for chatting. Have a great day!")

                    # Show analytics before exiting
                    self.show_analytics()

                    break

                # Generate chatbot response
                response = self.get_response(user_message)

                # Print chatbot response
                print(f"ChatBot: {response}\n")

            # Handling keyboard interruption
            except KeyboardInterrupt:

                print("\n\nChat interrupted by user.")
                break

            # Handling unexpected errors
            except Exception as e:

                print(f"\nError Occurred: {e}")

# RUNNING THE CHATBOT

# Creating chatbot object
bot = ChatBot()

# Starting chatbot
bot.chat()
```
